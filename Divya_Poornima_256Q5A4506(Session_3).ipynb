**Q1. Write a program using a for loop to print all even numbers from 1 to 20.**

```python
# Even numbers from 1 to 20
print("Even numbers from 1 to 20 is:")
for i in range(1, 21):
  if i%2==0:
    print(i)
```

**Q2. Write a while loop program to find the sum of the digits of a number entered by the user.**

```python
# take input value from user
k = int(input("Enter the Number: "))
sum = 0

# Use this abs function for convert negitave values to positive values
if k < 0:
  k = abs(k)

while k > 0:
  last_digit = k % 10      # it takes last digit
  sum += last_digit        # sum of last digits
  k //= 10                 # it remove last digit after added
print(sum)
```

**Q3. Using a for loop, count how many vowels are present in a word entered by the user.**

```python
# Find Vowels in Letter

letter = input("Enter the Word: ")
vowels = "aeiouAEIOU"                   # vowels for checking
count = 0                               # to count the vowels

for words in letter:                    # words to check each value in letter
  if words in vowels:                   # words to check each value in vowels
    count +=1
    print(words)
print(f"Number of Vowels in {letter}: {count}")
```

**Q4. Using do-while logic in Python (while True + break), repeatedly ask the user to enter a number until they type "stop".**

```python
# Repeatedly ask the user to enter a number
while True:                            # while true is for continues loop
  num = input("Enter a Number:")
  if num.upper() == "STOP":              # .upper is convert string values in upper case like stop into STOP
    break                              # brek is for break the loop
```

**Q5. Write a for loop program to print the multiplication table of a number entered by the user from 1 to 10.**

```python
# The multiplication table of a number entered by the user
a = int(input("Enter Table Number: "))      # it takes table number from user

for i in range(1,11):                       # the loop prints table step by step
  print(f"{a} * {i} = {a*i}")               # it display the table from 1 to 10
```
