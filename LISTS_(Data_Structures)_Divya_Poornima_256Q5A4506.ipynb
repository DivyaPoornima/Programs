***6.Write a program to update an element in a list using its index.***

* To update an element in a list using its index in Python, you can directly access the element using square bracket notation [] and assign a new value to it. This is a fundamental feature of lists in Python, as they are mutable.

```python
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")


index_to_update = 2
new_value = 99


my_list[index_to_update] = new_value


print(f"Updated list:   {my_list}")
```

***7.Write a program using list methods to:***

  python list methods are built-in functions that allow us to perform various operations on lists, such as adding, removing, or modifying elements.

*  append(element): This method adds a single element to the end of the list.
*  sort(): This method sorts the list in place (modifies the original list). By default, it sorts in ascending order.

*   reverse(): This method reverses the order of elements in the list in place.

```python
numbers = [40, 10, 30, 20]

numbers.append(50)
print("After append:", numbers)

numbers.sort()
print("After sort:", numbers)

numbers.reverse()
print("After reverse:", numbers)
```

***8. What is a nested list? Write a program to create a nested list and access a specific element***

  * A nested list is a list that contains
one or more other lists as its elements.

```python
nested_list = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]


print("Nested List:")
print(nested_list)


print("\nElement at row 2, column 3:")
print(nested_list[1][2])


print("\nAll elements:")
for row in nested_list:
    for item in row:
        print(item, end=" ")
    print()
```

***9.Write a program to implement a real-time application of lists (student attendance or shopping cart).***

***Features***



*   Add student
*   View students
*   Exit
*   Uses list + loop + menu

```python
students = []

while True:
    print("\n--- Student Application System ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        roll = input("Enter roll number: ")
        students.append([name, roll])
        print(" Student added successfully")

    elif choice == "2":
        if len(students) == 0:
            print(" No students found")
        else:
            print("\nStudent List:")
            for s in students:
                print("Name:", s[0], "| Roll:", s[1])

    elif choice == "3":
        print(" Exit")
        break

    else:
        print(" Invalid choice")
```

***10.Explain the time complexity of the following list operations******10.Explain the time complexity of the following list operations***

* **Traversal**

  *  Visiting each element in the list one byone.

          Time Complexity: O(n)

        * If there are n elements, you must check all n.

* ***searching***


*   Time Complexity: O(n)

     You may need to check every element until it is found.

* ***Insertion***

  1.   At the end (append)
      O(1)

  2.   At the beginning or middle
     O(n) (elements must be shifted)

*  **Deletion**

* Removing an element from the list.

1.   From the end (pop)
         O(1)
2.   From the beginning or middle
      O(n) (elements must be shifted)
