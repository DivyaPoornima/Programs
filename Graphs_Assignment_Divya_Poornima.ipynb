**Q1. Define a graph formally. Explain its components and mention any three applications of graphs.**

**Formal Definition:**
* A graph is a non-linear data structure.
* A graph G is defined as an ordered pair G=(V,E).

Where,

V=Set of Vertices (nodes)

E=Set of Edges (connections b/w vertices)

**Components of a Graph:**
* **Vertices:** It represents entities.
* **Edges:** It represents connections between vertices.
* **Degree:** Number of edges connected to a vertex.
* **Directions/Weights:** Some graphs have arrows means directions and weights means cost or distance.

**Applications of Graphs:**
* **Social Networks:** Represent people as nodes and friendships as edges.
* **Google Maps:** Cities as nodes, roads as with weights(distance).
* **Computer Networks:** Computers as nodes and connections are edges.

**Q2. Differentiate between:**

**(a) Directed and Undirected Graphs**
**(b) Weighted and Unweighted Graphs**
**(with definitions and examples)**

**a)** **Directed and Undirected Graph:**
*  **Directed Graph:**

A graph where edges have a specific directions is known as directed graph. It shown using arrows(->).

* **Undirected Graph:**

A graph where edges have no directions is known as undirection graphs. The connections are two ways.


**b) Weighted and Unweighted Graphs:**
* **Weighted Graph:**

A graph where each edge has a weight(cost, distance, time, etc..) is known as weighted graph.
* **Unweighted Graph:**

A graph where edges do not have any weights is known as unweighted graph.

**Q3. What are Cyclic and Acyclic graphs? Explain their effect on graph traversal and problem-solving.**

**Cyclic Graph:**
* A graph that contains atleast one cycle is known as cyclic graph.
* Cyclic Graphs can cause infinite loops if traversal is not carefully managed.

**Acyclic Graph:**
* A graph that have no cycles is known as acyclic graph.
* Acyclic Graphs are easier for algorithms such as dependency resolution or topological sorting.

**Q4. Define Connected and Disconnected graphs. Explain their importance in real-world applications.**

**Connected Graph:**
* A graph where every vertex is reachable from every other vertex.
* It ensures communications, transport, data flow is possible.
* **EX:** A Network where all computers are linked.

**Disconnected Graph:**
* A graph where some vertices cannot be reached from other vertices.
* It highlights isolated systems or communities, useful in clustering and networking analysis.
* **EX:** Two separate social groups with no friendship between them.

**Q1. Write a Python program to create an undirected unweighted graph using an adjacency list.**

**Add at least 4 vertices and 4 edges**

**Display the adjacency list**

```python
graph={
    "A":["B", "C"],
    "B":["A", "D"],
    "C":["A", "D"],
    "D":["B", "C"]
}
print("Adjacency List:")
for vertex in graph:
    print(vertex, ":", graph[vertex])
```

**Q2. Write a Python program to represent a graph using an adjacency matrix.**

**Initialize the matrix**

**Add edges between given vertices**

**Print the matrix**

```python
vertices=["A", "B", "C", "D"]
n=len(vertices)
matrix=[[0]*n for _ in range(n)]
edges=[("A", "B"), ("A", "C"), ("B", "D"), ("C", "D")]
for u, v in edges:
  i, j=vertices.index(u), vertices.index(v)
  matrix[i][j]=1
  matrix[j][i]=1
print("Adjacency Matrix Representation:")
for row in matrix:
  print(row)
```

**Q4. Write a Python program to check whether a given graph is connected or disconnected using graph traversal logic (DFS or BFS).**

```python
graph={
    "A":["B", "C"],
    "B":["A", "D"],
    "C":["A", "D"],
    "D":["B", "C"]
}
visited=set()
def dfs(node):
  if node not in visited:
    visited.add(node)
    for neighbour in graph[node]:
      dfs(neighbour)
start=list(graph.keys())[0]
dfs(start)
if len(visited)==len(graph):
  print("The graph is connected.")
else:
  print("The graph is disconnected.")
```
