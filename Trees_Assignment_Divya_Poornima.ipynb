**Q1. What is a Tree data structure?Why is a tree called a non‑linear data structure?**


* A tree is a non-linear data structure.
* It made up of nodes connected by edges. The tree orgnaize data in a branching structure.
* The starting single node is called a Root nodes and the branches of the root is called Child nodes.
* The tree is called non-linear and hierarchical data structures because unlike arrays, linked lists (which store data sequentially).
* The elements are not stored one after another but in levels, making it easier to represent hierarchical relationships.

**Q3.Define the following tree terminologies:**

**Node, Root, Parent, Child, Leaf, Level, Height**

* **Node-** A node is called a single element in a tree.
* **Root-** Root means the topmost node of a tree.
* **Parent-** A node that has branches (children) leading to other nodes.
* **Child-** A node that comes directly under a parent.
* **Leaf-** A node without any children or end of a branch.
* **Level-** The depth of a node in the tree like root is level 0, its children are level 1 and etc...
* **Height-** A path from the root node to leaf node.

**Q6. What is a Binary Tree?**
**Why can a binary tree have at most two children?**

* A Binary tree is a tree where each node have 0, 1 or 2 children.

* It can have at most two children because one child is placed on the left and the other on the right.

* This makes traversal and searching algorithms simpler.

**Q7.Explain the Binary Search Tree (BST).Write the BST property rules.**

* A Binary Search Tree is a special type of binary tree where nodes are arranged in an ordered way.

* It is used for fast searching.

**BST Property Rules:**
1. Left child must smaller than Parent node.
2. Right child must grater than parent node.
3. Both left and right subtrees must also follow this rule recursively.

* This property ensures efficient searching, insertion and deletion.

**Q9. What is Tree Traversal?**
**Explain the following traversal techniques:**

**Inorder, Preorder, Postorder, Level Order**

Visiting all nodes of a tree in a specific order.

**Techiniques:**
* **Inorder:** (Left-> Root-> Right)

Useful for BSTs because it gives sorted order.
* **Preorder:** (Root-> Left-> Right)

Used to create a copy of the tree.
* **Postorder:** (Left-> Right-> Root)

Useful for deleting/freeing nodes.
* **Level Order:**

Visits nodes level by level like breadth-first search.

**Q10. Write a Python program to:**

**Create a binary tree node**

**Perform Inorder Traversal**

**Print the output**

```python
class Node:
  def __init__(self, data):
    self.data=data
    self.left=None
    self.right=None
def inorder_traversal(root):
  if root:
    inorder_traversal(root.left)
    print(root.data, end=" ")
    inorder_traversal(root.right)

root=Node(10)
root.left=Node(5)
root.right=Node(15)
root.left.left=Node(2)
root.left.right=Node(7)
root.right.left=Node(12)
root.right.right=Node(20)

print("Inorder Traversal Output:")
inorder_traversal(root)
```
