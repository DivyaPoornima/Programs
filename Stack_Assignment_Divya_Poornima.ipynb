**Q1. What is a Stack? Explain the LIFO (Last In First Out) principle with a real‑life example.**

A Stack is a Linear Data structure that stores elements in an ordered manner.

It follows LIFO (Lastb In First Out) principle.It means the last inserted element is removed first.

**Ex: Bangles**

When we need to remove the bangles from hand, we remove the last inserted bangle first.

**Q3. Write the difference between Stack and Queue.**

**Stack:**

1.   Stack is a Linear data structure.
2.   It follows LIFO (Last In First Out) principle.
3.   The Operations we use in stack are Push, pop, peek/top.

**Queue:**

1.   A Queue is a Linear data structure.
2.   It follows FIFO (First In First Out) principle.
3.   The Operations we use in Queue are Enqueue, Dequeue, Front.





**Q4. What is stack underflow?**
**When does it occur.**

Stack Underflow is an error condition that occurs when we try to remove an element from an empty stack.

It happens beacause there are no elements left to remove.

**Q5. Stack Creation**

```python
stack=[]                   #empty stack
for i in range(1, 6):      #push five elements
  stack.append(i)
print("Stack after pushing 5 elements:", stack)   #display stack
```

**Q6. Push Operation**

```python
stack=[]
element=input("Enter the element to be push: ")
stack.append(element)
print("Stack after pushing the element:", stack)
```

**Q7. Pop Operation**

```python
stack=[10, 20, 30]
if len(stack)==0:
  print("stack is empty! Cannot pop.")
else:
  popped=stack.pop()
  print("Popped element:", popped)
  print("Updated Stack:", stack)
```

**Q8. Peek Operation**

```python
stack=[5, 10, 15]
if len(stack)==0:
  print("stack is empty! Nothing to peek.")
else:
  print("Top element of stack:", stack[-1])
```

**Q9. Check Stack Status**

```python
stack=[1, 2, 3, 4]
if len(stack)==0:
  print("Stack is empty.")
else:
  print("Stack is not empty.")
  print("Number of elements in stack:", len(stack))
```

**PART-C: APPLICATION‑BASED QUESTION**

**Q10. Stack Using Functions**

```python
stack=[]
def push(element):
  stack.append(element)
  print(f"Stack after pushing the element: {element}")

def pop():
  if len(stack)==0:
    print("Stack is empty! Cannot pop.")
  else:
    removed=stack.pop()
    print(f"Popped element: {removed}")

def peek():
  if len(stack)==0:
    print("Stack is empty! Cannot peek.")
  else:
    print(f"Top element: {stack[-1]}")

push(5)
push(16)
peek()
pop()
```
