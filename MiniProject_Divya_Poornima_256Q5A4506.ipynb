```python
# PART 1:Welcome Screen
print("-----------------------------------------")
print("     STUDENT PROGRESS TRACKER SYSTEM     ")
print("-----------------------------------------\n")

# Ask the user to enter Details
name = input("Enter Student Name: ")
student_class = input("Enter Student Class: ")
rollno = input("Enter Student Roll Number: ")
.
# PART 2: Enter Subject Marks
def students_data():
  marks = []

  # taking subjects list from user
  subs = int(input("\nHow many Subjects do you want to enter: "))
  totalmarks = 0
  print("\n")

  # taking marks of subjects from user
  for i in range(subs):
    while True:
        mark = int(input(f"Enter Marks for Subject {i+1}: "))
        if 0 <= mark <= 100:
            marks.append(mark)
            totalmarks += mark
            break
        else:
            print("Enter valid marks (0–100)")

  # Calsulation by using marks
  averagemarks = totalmarks / subs                   # Average Marks
  highest = max(marks)                               # Highest Marks
  lowest = min(marks)                                # Lowest Marks
  percentage = (totalmarks / (subs * 100)) * 100     # Percentage of the student

  # PART 3: Grade Calculations
  if averagemarks >= 90:                             # 90 and Above A Grade
   grade = "A"
  elif averagemarks >= 75:                           # 75 - 89  B Grade
   grade = "B"
  elif averagemarks >= 50:                           # 50 - 74  C Grade
   grade = "C"
  elif averagemarks >= 35:                           # 35 - 49  D Grade
   grade = "D"
  else:                                              # Below 35 Fail
   grade = "Fail"

  #Remarks (Excellent / Good / Needs Improvement)
  if grade == "A":
   remarks = "Excellent"
  elif grade == "B":
   remarks = "Good"
  else:
   remarks = "Needs Improvement"

  # PART 4: Student Pass/Fail Logic
  status = "Pass"
  for mark in marks:
    if mark < 35:
      status = "Fail"
      break

  # The student failed because he failed one subject.
  if status == "Fail":
    grade = "Fail"
    remarks = "Needs Improvement"
  else:
    status = "Pass"

  print("\n")
  return totalmarks, averagemarks, highest, lowest, percentage, grade, status, remarks

totalmarks, averagemarks, highest, lowest, percentage, grade, status, remarks = students_data()

# PART 5: Display Progress Report
while True:
  print("-----------------------------------------")
  print("     STUDENT PROGRESS TRACKER SYSTEM     ")
  print("-----------------------------------------")
  print(f"Class         : {student_class.upper()}")
  print(f"Roll Number   : {rollno}")
  print(f"Highest Marks : {highest}")
  print(f"Lowest Marks  : {lowest}")
  print(f"Total Marks   : {totalmarks}")
  print(f"Average Marks : {averagemarks}")
  print(f"Percentage    : {percentage}%")
  print(f"Grade         : {grade}")
  print(f"Result        : {status}")
  print(f"Remarks       : {remarks}")

  # PART 6: Menu
  print("\nWhat do you want to do next?\n")
  print("1. Enter marks again")
  print("2. View report again")
  print("3. Exit\n")

  # User input for choices
  k = input("Enter your choice: ")

  if k == "1":
    totalmarks, averagemarks, highest, lowest, percentage, grade, status, remarks = students_data()
  elif k == "2":
    continue
  elif k == "3":
    print("Exiting... Thank You!")
    break
  else:
    print("Enter valid choice again.\n")
```
