**Introduction to Variables**

What is a Variable?

A variable is a name that refers to a value stored in memory.

**String**

A string stores text.
Real-life example: a user’s name in an app.

```python
#Variables & Types (Python automatically assign the data types)
name = "Divya"         #string

print(f"My Name is {name} and It is a {type(name)}")     #f-strings gives the clear and clean format output
```

**Float**


A float represents a number with decimals.
Real-life example: the price of a product.

```python
salary = 2.5             #float

print(f"My Salary Package is {salary}LPA and It is a {type(salary)}")
```

**Integer**

Integer is a whole number- a number without a decimal point.

```python
num = 20                 #integer

print(f"The given Number is {num} and It is a {type(num)}")                   #type prints data type of variable
```

**Boolean**

Boolean values are either True or False.
Real-life example:

Is the user logged in? → Yes/No

Is the battery low? → True/False

```python
is_age_above_18 = True   #boolean

print(f"Is my age is above 18: {is_age_above_18} \n It is a {type(is_age_above_18)}")
```

**Q3-Simple Calculator**

```python
# Taking two numbers as input
#Simple Calculator
num1 = int(input("Enter a 1st Number:"))
num2 = int(input("Enter a 2nd Number:"))

add = num1 + num2          #add is storing a addition values
sub = num1 - num2          #sub is storing a subtration values
mul = num1 * num2          #mul is storing a multiplication
div = num1 / num2          #div is storing a division

print(f"Addition of given values are:{add}")
print(f"Subtraction of given values are:{sub}")
print(f"Multiplication of given values are:{mul}")
print(f"Division of given values are:{div}")
```

**Operators**

**Q4-Even/Odd or Positive/Negative**

```python
num = int(input("Enter a Number:"))

if num > 0:                                    #here, we use nested if for check num > 0  then check positive and negitive
  print(f"{num} is a Positive Number.")
  if num % 2 == 0:
    print(f"{num} is a Even Number.")
  else:
    print(f"{num} is a Odd Number.")
elif num < 0:
  print(f"{num} is a Negitive Number.")
else:
  print("The number is Zero")
```

**Q5-Comparison Operator**

```python
#user entered data is must in integer
num1 = int(input("Enter a first Number:"))
num2 = int(input("Enter a second Number:"))

if num1 == num2:
  print("Both Numbers are Equal")
elif num1 > num2:
  print(f"{num1} is Graterthan {num2}")
else:
  print(f"{num1} is Lessthan {num2}")
```

**Q7-Grading System**

```python
marks = int(input("Enter a marks between 0 to 100:"))

if  marks >= 0 and marks <= 100:
  if marks >= 90:
    Grade = "A"
  elif marks >=75:
    Grade = "B"
  elif marks >= 50:
    Grade = "C"
  elif marks >= 35:
    Grade = "D"
  else:
    print("Fail")
  print(f"Your marks is {marks}, You got {Grade} Grade")
else:
  print("Enter marks between 0 to 100 only")
```

**Q8-Movie Ticket Pricing**

```python
age = int(input("Enter your age:"))

if age < 5:
  price = "Free"
elif age <= 17:
  price = "₹100"
else:
    price = "₹150"

print(f"Your Ticket Price is {price}")
```
